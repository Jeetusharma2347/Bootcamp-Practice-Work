## Data import for visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
df.duplicated().sum()
df_clean=df.drop_duplicates(inplace=True)
```

```python
print(df["continent"].value_counts())

continent_summary = df.groupby("continent")[["lifeExp","gdpPercap","pop"]].mean()

print(continent_summary)
```

```python
print("Year range:", df["year"].min(), "to", df["year"].max())

print("Number of countries:", df["country"].nunique())

print("Number of continents:", df["continent"].nunique())

print("Average life expectancy:", df["lifeExp"].mean())

print("Minimum life expectancy:", df["lifeExp"].min())

print("Maximum life expectancy:", df["lifeExp"].max())

print(df.describe())
```

```python
#using the lib matplotlib #I need a bar chart that shows me avg life expectancy by continent
continent_life=df.groupby("continent")["lifeExp"].mean().sort_values()
plt.figure(figsize=(8, 5))
plt.bar(continent_life.index, continent_life.values)
plt.title(f"Average Life Expectancy by Continent in (latest_year)")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Line chart life expectancy over time for 3 countries India, China, United States
selected_countries = ["India", "China", "United States"]

trend_df = df[df["country"].isin(selected_countries)]

plt.figure(figsize=(9, 5))

for country in selected_countries:
    country_df = trend_df[trend_df["country"] == country]
    plt.plot(country_df["year"],country_df["lifeExp"],marker="o",label=country
    )

plt.title("Life Expectancy Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
